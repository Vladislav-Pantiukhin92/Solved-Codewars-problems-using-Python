### Calculate Fibonacci return count of digit occurrences - 5 kyu, Codewars/Python

```python
"""
Fibonacci numbers are generated by setting F0 = 0, F1 = 1, and then using the formula:
Fn = Fn-1 + Fn-2
Your task is to efficiently calculate the nth element in the Fibonacci sequence and then count the 
occurrence of each digit in the number. Return a list of integer pairs sorted in descending order.
10 ≤ n ≤ 100000

Examples
f(10) = 55  # returns [(2, 5)], as there are two occurances of digit 5
f(10000)    # returns:
  [(254, 3),
   (228, 2),
   (217, 6),
   (217, 0),
   (202, 5),
   (199, 1),
   (198, 7),
   (197, 8),
   (194, 4),
   (184, 9)]
If two integers have the same count, sort them in descending order.
Your algorithm must be efficient.
"""

from collections import Counter # 'collections.Counter' will be used to create the dictionary with numbers as keys 
# and numbers' occurrences as values

# To calculate the specific fibonacci number we are going to utilize the Dynamic Programming (bottom-up) approach - 
# tabulation. We will build the array with fibonacci numbers, where our relevant sought-after number is the 
# last element within the aforementioned array:

def fib_digits(n):
    my_list = [0, 1] # Two starting fibonacci positions:
    for i in range(2, n + 1): # Generation of the fibonacci sequence
        my_list.append(0)
        my_list[i] = my_list[i - 1] + my_list[i - 2]
    # Right now we have the relevant fibonacci number - the last element within the aforementioned array
    # The next step is to create the dictionary and separately extract its keys and values, 
    # using the dictionary methods:
    numbers = list(map(int, dict(Counter(str(my_list[-1]))).keys())) # Extraction of keys, using dict.keys()
    occurrence = list(map(int, dict(Counter(str(my_list[-1]))).values())) # Extraction of values, using dict.values()
    # The next step is to merge keys and values together (using the 'zip' function) 
    # and obtain the list with iterables (sorted in descending order):
    final = sorted(list(zip(occurrence, numbers)), reverse=True)
    return final # The list with iterables is returned

fib_digits(10000)
```
